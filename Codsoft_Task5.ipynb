```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
#Linking Dataset to Colab
import io
data = pd.read_csv((io.BytesIO(uploaded["creditcard.csv"])), encoding ='unicode_escape')
```

```python
legit=data[data.Class==0]
fraud=data[data.Class==1]
```



*   **UNDER SAMPLING**



```python
sample=legit.sample(n=492)
#concatenating sample and fraud series to get our new balanced dataset
```

```python
new_data=pd.concat([sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```



*   **SPLITTING THE DATA INTO FEATURES & TARGET**


```python
X=new_data.drop(columns='Class',axis=1)
Y=new_data['Class']
```

```python
X
```

```python
Y
```



*   **SPLIT THE DATA INTO TRAINING & TESTING DATA**



```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```



*   **MODEL TRAINING - LOGISTIC REGRESSION**



```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```



*   **MODEL EVALUATION - ACCURACY SCORE**



```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on Training data:",training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on Testing data:",test_data_accuracy)
```
